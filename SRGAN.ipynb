```python
import cv2
import numpy as np
import os
import matplotlib.pyplot as plt
```

```python
train_dir = 'D:/Desktop/Improv/Formal Coding/Jupyter Notebooks/Minor Project/SRDataset/flickr30k_images'
i=0
n=5000
train_dir
```

```python
for img in os.listdir( train_dir + '/flickr30k_images' )[:n]:
    print(img)
#     img_array = cv2.imread(train_dir +'/flickr30k_images'+ img)
    fname = 'flickr30k_images'
    img_array = cv2.imread((os.path.join(train_dir,fname, img)))
    img_array = cv2.resize(img_array, (128,128))
    lr_img_array = cv2.resize(img_array,(32,32))
    hrop = "hr_5000"
    lrop = "lr_5000"
    cv2.imwrite((os.path.join(train_dir, hrop, img)), img_array)
    cv2.imwrite((os.path.join(train_dir, lrop, img)), lr_img_array)
    i=i+1
    print(i)
```
